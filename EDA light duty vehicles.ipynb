```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ldv = pd.read_csv('Light Duty Vehicles.csv')
ldv.head(2)
```

```python
ldv.info()
```

```python
ldv.describe()
```

```python
# Muestra la distribucion de los tipos de combustible
ldv['Fuel'].value_counts().plot(kind='bar')
plt.title('Distribucion de los tipos de combustible')
plt.show()
```

```python
# Filtrar el DataFrame para incluir solo las filas donde 'Conventional Fuel Economy Combined' sea mayor que 0
ldv_filtered = ldv[ldv['Conventional Fuel Economy Combined'] > 0]

# Calcular la media del consumo de combustible combinado por tipo de combustible
fuel_consumption_mean = ldv_filtered.groupby('Fuel')['Conventional Fuel Economy Combined'].mean()

# Crear el grÃ¡fico de barras con los valores filtrados
fuel_consumption_mean.plot(kind='bar')
plt.title('Consumo promedio de combustible por tipo de combustible')
plt.show()
```
