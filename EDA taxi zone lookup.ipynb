```python
import pandas as pd
import numpy as np

tzl = pd.read_csv('taxi+_zone_lookup.csv')
tzl.head(2)
```

```python
tzl.info()
```

```python
tzl.describe()
```

```python
unique_values_borough = tzl['Borough'].value_counts()
unique_values_zone = tzl['Zone'].value_counts()
unique_values_service_zone = tzl['service_zone'].value_counts()

print("Valores únicos en la columna 'Borough':")
print(unique_values_borough)
print("\nValores únicos en la columna 'Zone':")
print(unique_values_zone)
print("\nValores únicos en la columna 'service_zone':")
print(unique_values_service_zone)
```

```python
import matplotlib.pyplot as plt

# Configuración de tamaño de figura
plt.figure(figsize=(10, 6))

# Gráfico de barras para la columna 'Borough'
plt.subplot(1, 3, 1)
tzl['Borough'].value_counts().head(5).plot(kind='bar', color='skyblue')
plt.title('Distribución de Distritos')
plt.xlabel('Distritos')
plt.ylabel('Frecuencia')

# Gráfico de barras para la columna 'Zone'
plt.subplot(1, 3, 2)
tzl['Zone'].value_counts().head(10).plot(kind='bar', color='salmon')
plt.title('Distribución de las Zonas (Primeros 10)')
plt.xlabel('Zona')
plt.ylabel('Frecuencia')

# Gráfico de barras para la columna 'service_zone'
plt.subplot(1, 3, 3)
tzl['service_zone'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribución de las zonas de servicio')
plt.xlabel('zona de servicio')
plt.ylabel('Frecuencia')

# Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```
